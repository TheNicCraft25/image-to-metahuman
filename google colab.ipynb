```python
import os

# Verzeichnis, in dem sich die Dateien befinden
verzeichnis = '/content/Deep3DFaceRecon_pytorch/input'

# Alle Dateien im Verzeichnis auflisten
dateien = os.listdir(verzeichnis)

# Schleife durch jede Datei im Verzeichnis
for datei in dateien:
    # Den vollständigen Pfad zu jeder Datei erstellen
    alter_pfad = os.path.join(verzeichnis, datei)

    # Überprüfen, ob die Datei kein Ordner ist
    if os.path.isfile(alter_pfad):

        basisname, erweiterung = os.path.splitext(datei)

        # Den neuen Dateinamen erstellen (hier als Beispiel: füge "_neu" hinzu)
        neuer_name = 'input' + erweiterung

        # Den neuen Pfad für die umbenannte Datei erstellen
        neuer_pfad = os.path.join(verzeichnis, neuer_name)

        # Die Datei umbenennen
        os.rename(alter_pfad, neuer_pfad)
```

```python
%cd /content/Deep3DFaceRecon_pytorch
!source activate deep3d_pytorch

import face_alignment
import numpy as np
import os
#from skimage import io

fa = face_alignment.FaceAlignment(face_alignment.LandmarksType.TWO_D, flip_input=False, device='cpu', face_detector='sfd')

preds = fa.get_landmarks_from_directory('input')
#print(preds)

for i in preds:
    file_name = os.path.split(i)[1].split(".")[0]
    print(file_name)
    out = preds[i][0]

    Lm3D = out
    lm_idx = np.array([31,37,40,43,46,49,55]) - 1
    Lm3D = np.stack([Lm3D[lm_idx[0],:],np.mean(Lm3D[lm_idx[[1,2]],:],0),np.mean(Lm3D[lm_idx[[3,4]],:],0),Lm3D[lm_idx[5],:],Lm3D[lm_idx[6],:]], axis = 0)
    Lm3D = Lm3D[[1,2,0,3,4],:]
    content = str(Lm3D).replace('[', '').replace(']', '')
    print(content)
    f = open('/content/Deep3DFaceRecon_pytorch/input/detections/' + file_name + '.txt', "w")
    f.write(content)
    f.close()
```
